```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0, 10)
y = 2 * x ** 2 - 3 * x + 1
plt.plot(x, y, \
    color = "green", ls = "dashed", lw = 3, \
    marker = "o", mfc = "blue", ms = 12)
```

```python
x = np.arange(0, 10)
y = 2 * x ** 2 - 3 * x + 1
plt.plot(x, y, \
    color = (0, 0, 0.7), ls = "dotted", lw = 3, \
    marker = "<", mfc = "yellow", mec = "#800000", \
    ms = 10)
```

```python
x = np.arange(0, 10)
y = 2 * x ** 2 - 3 * x + 1
plt.plot(x, y, ls = "-.", marker = ".", ms = 10)
```

```python
x = np.random.randint(0, 100, 20)
y = np.random.randint(0, 100, 20)
plt.scatter(x, y, c = "red", s = 20, marker = ">")
```

```python
x = np.random.randint(0, 100, 20)
y = np.random.randint(0, 100, 20)
c = np.random.randint(0, 20, 20)
s = np.random.randint(0, 1000, 20)
plt.scatter(x, y, c = c, s = s, cmap = plt.cm.RdYlGn, alpha = 0.5)
plt.colorbar()
```

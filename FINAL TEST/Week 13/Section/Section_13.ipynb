```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0, 10, 0.1)
y = 2 * x ** 2 - 3 * x + 1
plt.plot(x, y)
```

```python
x = np.arange(0, 10)
y = 2 * x ** 2 - 3 * x + 1
plt.plot(x, y, 'o:r')
```

```python
x = np.random.randint(0, 100, 200)
y = np.random.randint(0, 100, 200)
plt.scatter(x, y)
```

```python
x = np.arange(1, 6)
y = [10, 5, 25, 3, 22]
plt.bar(x, y)
```

```python
x = np.arange(1, 6)
y = [10, 5, 25, 3, 22]
plt.barh(x, y)
```
